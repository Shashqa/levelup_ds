```python
from math import sqrt

class Shape():
    # The class Shape contains information about geometric shapes
    
    def __init__(self, ang, center_x = 0, center_y = 0):
        # Each shape is characterized by the number of angles 
        # and the coordinates of the center
        self.ang = ang
        self.center_x = center_x
        self.center_y = center_y

    def shape_name(self):
        # The shapes have their own name depending on the number of angles
        names = {0:'circle', 3:'triangle', 4:'quadrangle', 5:'pentagon', 6:'hexagon', 
                 7:'heptagon', 8:'octagon', 9:'nonagon', 10:'decagon'}
        if self.ang in names.keys():
            name = names[self.ang]
        elif self.ang > 10:
            name = 'polygon'
        else:
            name = "doesn't exist" 
        print ('This geometric shape -', name)
     
    def sum_ang(self):
        # The sum of the inner corners of a polygon depends on the number 
        # of its corners. Method allows to calculate this amount.
        if self.ang >= 3:
            summ = 180*(self.ang - 2)
        else:
            summ = 0
        return summ
    
    def move_center(self, increment_x, increment_y):
        # The method changes the coordinates of the center of 
        # the shape by a given number
        self.center_x += increment_x
        self.center_y += increment_y
    
class Triangles (Shape):
    # The class Triangles contains information for working with triangles
    
    def __init__(self, ang, a, b, c, center_x = 0, center_y = 0):
        # Each triangle is characterized by the lengths of its sides
        super().__init__(ang, center_x = 0, center_y = 0)
        self.a = a
        self.b = b
        self.c = c
        
    def check_existence(self):
        # A triangle exists if the sum of the two sides of the triangle 
        # is greater than the third side.
        sides = [self.a, self.b, self.c]
        for i in range(len(sides)):
            if (sides[i] >= sum(sides) - sides[i]) or self.ang != 3:
                existence = False
                break
            else:
                existence = True
            i += 1
        return existence
    
    def calculate_area(self):
        # The area of a triangle can be expressed in terms of the lengths 
        # of its three sides using Heron's formula
        p = (self.a + self.b + self.c)/2
        area = sqrt(abs(p*(p - self.a)*(p - self.b)*(p - self.c)))
        return area
    
    def check_similarity(self, other_triangle):
        # Triangles are similar if the three sides of one triangle are 
        # proportional to the three sides of the other
        if self.a/other_triangle.a == self.b/other_triangle.b == self.c/other_triangle.c:
            k_sim = self.a/other_triangle.a
            similarity = True
        else:
            similarity = False
            k_sim = 0
        return similarity, k_sim
    
abc = Triangles(3,1,2,2.4)
abc.shape_name()

ex = abc.check_existence()
print ('This triangle%s exist%s' %(" doesn't"*(not ex), 's'*ex))

if ex == True:
    print ('The area of the triangle is %.2f cm2' %abc.calculate_area())

mnk = Triangles(3,3,6,7.2)
smlr = abc.check_similarity (mnk)
print ('The triangles are%s similar.%s %.2f' %(' not'*(not smlr[0]), 'The similarity coefficient is', smlr[1]))
    
```
